```python
import os
import pickle
import numpy as np
import pandas as pd
from scipy import stats
from scipy.signal import butter, filtfilt, firwin, savgol_filter
import warnings
warnings.filterwarnings('ignore')

print(" BẮT ĐẦU PHASE 1: LỌC TÍN HIỆU VÀ TRÍCH XUẤT ĐẶC TRƯNG...\n")

# =====================================================================
# BƯỚC 1: XÂY DỰNG BỘ LỌC TÍN HIỆU ĐÚNG CHUẨN BÀI BÁO SELF-CARE
# =====================================================================
def apply_filters_wrist(acc, bvp, eda, temp):
    # 1. Lọc ACC (Gia tốc - 32Hz): Dùng bộ lọc FIR (chiều dài 64, cắt ở 0.4 Hz)
    fir_coeff = firwin(64, 0.4, fs=32)
    acc_x = filtfilt(fir_coeff, 1.0, acc[:, 0])
    acc_y = filtfilt(fir_coeff, 1.0, acc[:, 1])
    acc_z = filtfilt(fir_coeff, 1.0, acc[:, 2])
    acc_filtered = np.column_stack((acc_x, acc_y, acc_z))

    # 2. Lọc BVP (Nhịp tim - 64Hz): Dùng Butterworth band-pass (0.7 - 3.7 Hz), bậc 3
    # Lọc lấy dải nhịp tim chuẩn của người (40 - 220 BPM)
    b_bvp, a_bvp = butter(3, [0.7, 3.7], btype='bandpass', fs=64)
    bvp_filtered = filtfilt(b_bvp, a_bvp, bvp.flatten())

    # 3. Lọc EDA (Điện da - 4Hz): Dùng Butterworth low-pass (1 Hz), bậc 6
    b_eda, a_eda = butter(6, 1.0, btype='low', fs=4)
    eda_filtered = filtfilt(b_eda, a_eda, eda.flatten())

    # 4. Lọc TEMP (Nhiệt độ - 4Hz): Dùng Savitzky-Golay filter (window=11, order=3)
    temp_filtered = savgol_filter(temp.flatten(), window_length=11, polyorder=3)

    return acc_filtered, bvp_filtered, eda_filtered, temp_filtered

# =====================================================================
# BƯỚC 2: HÀM CẮT CỬA SỔ & TRÍCH XUẤT ĐẶC TRƯNG
# =====================================================================
def process_single_subject(file_path, subject_id):
    with open(file_path, 'rb') as file:
        data = pickle.load(file, encoding='latin1')

    # Lấy dữ liệu thô từ cổ tay (Wrist)
    wrist = data['signal']['wrist']
    acc_raw, bvp_raw = wrist['ACC'], wrist['BVP']
    eda_raw, temp_raw = wrist['EDA'], wrist['TEMP']
    labels = data['label']

    # Áp dụng bộ lọc (Khử nhiễu)
    acc, bvp, eda, temp = apply_filters_wrist(acc_raw, bvp_raw, eda_raw, temp_raw)

    # Thông số tần số lấy mẫu (Hz)
    fs_eda, fs_temp, fs_bvp, fs_acc, fs_label = 4, 4, 64, 32, 700

    # Thông số cắt cửa sổ theo yêu cầu nhóm trưởng
    window_physio, window_acc, shift = 60, 5, 0.25
    shift_eda = int(shift * fs_eda) # Trượt 1 điểm (0.25 * 4 = 1)
    window_eda_samples = window_physio * fs_eda
    window_bvp_samples = window_physio * fs_bvp
    window_acc_samples = window_acc * fs_acc
    window_label_samples = window_physio * fs_label

    features = []
    total_steps = len(eda) - window_eda_samples

    for i in range(0, total_steps, shift_eda):
        t_start = i / fs_eda

        # Cắt các mảng (Segment)
        eda_w = eda[i : i + window_eda_samples]
        temp_w = temp[i : i + window_eda_samples]
        bvp_w = bvp[int(t_start * fs_bvp) : int(t_start * fs_bvp) + window_bvp_samples]

        # ACC lấy 5 giây cuối
        end_time_acc = t_start + window_physio
        start_time_acc = max(0, end_time_acc - window_acc)
        acc_w = acc[int(start_time_acc * fs_acc) : int(end_time_acc * fs_acc)]

        # Lấy nhãn
        label_w = labels[int(t_start * fs_label) : int(t_start * fs_label) + window_label_samples]
        if len(label_w) == 0: continue
        majority_label = stats.mode(label_w, keepdims=True)[0][0]

        # Bỏ qua dữ liệu nhiễu (chỉ lấy 1=Baseline, 2=Stress, 3=Amusement)
        if majority_label not in [1, 2, 3]:
            continue

        acc_mag = np.sqrt(acc_w[:,0]**2 + acc_w[:,1]**2 + acc_w[:,2]**2)

        # ==================================================
        # TRÍCH XUẤT ĐẶC TRƯNG HẠNG NẶNG (BÁM SÁT BẢNG 1 CỦA BÀI BÁO)
        # ==================================================
        features.append({
            'Subject_ID': subject_id,

            # --- Nhánh 1: BVP (Nhịp tim) ---
            'BVP_mean': np.mean(bvp_w),
            'BVP_std': np.std(bvp_w),
            'BVP_rms': np.sqrt(np.mean(bvp_w**2)),
            'BVP_dyn_range': np.max(bvp_w) - np.min(bvp_w),

            # --- Nhánh 2: EDA (Điện da/Mồ hôi) ---
            'EDA_mean': np.mean(eda_w),
            'EDA_std': np.std(eda_w),
            'EDA_min': np.min(eda_w),
            'EDA_max': np.max(eda_w),
            'EDA_dyn_range': np.max(eda_w) - np.min(eda_w),
            'EDA_slope': np.polyfit(np.arange(len(eda_w)), eda_w, 1)[0],

            # --- Nhánh 3: TEMP (Nhiệt độ) ---
            'TEMP_mean': np.mean(temp_w),
            'TEMP_std': np.std(temp_w),
            'TEMP_min': np.min(temp_w),
            'TEMP_max': np.max(temp_w),
            'TEMP_dyn_range': np.max(temp_w) - np.min(temp_w),
            'TEMP_slope': np.polyfit(np.arange(len(temp_w)), temp_w, 1)[0],

            # --- Nhánh 4: ACC (Vận động) ---
            'ACC_mean': np.mean(acc_mag),
            'ACC_std': np.std(acc_mag),
            'ACC_integral': np.sum(np.abs(acc_mag)), # Tích phân độ lớn chuyển động

            'Label_Original': majority_label
        })

    return pd.DataFrame(features)

print(" Đã khởi tạo xong các hàm Lọc tín hiệu & Trích xuất đặc trưng cho Phase 1!")
```

```python
# =====================================================================
# CELL 3: THỰC THI PHASE 1 - GOM DỮ LIỆU TỪ 15 NGƯỜI
# =====================================================================
base_folder = '/content/WESAD_Data/WESAD'
all_subjects_df = []

print(" Đang cày dữ liệu Phase 1... (Chỉ mất khoảng 1-2 phút)")

# Ghi đè lại shift = 5 (giây) vào trong hàm cũ để tối ưu tốc độ
def process_single_subject_v2(file_path, subject_id):
    global shift
    shift = 5 # Trượt 5 giây theo đúng chuẩn bài báo
    return process_single_subject(file_path, subject_id)

for root, dirs, files in os.walk(base_folder):
    for file in files:
        if file.endswith('.pkl'):
            subject_id = file.split('.')[0]
            print(f"  -> Trích xuất: {subject_id}")
            df_subject = process_single_subject(os.path.join(root, file), subject_id)
            all_subjects_df.append(df_subject)

df_master = pd.concat(all_subjects_df, ignore_index=True)
print(f"\n Đã gom xong Phase 1! Tổng dữ liệu có {df_master.shape[0]} cửa sổ.")

# Chuẩn bị Task 2-class và 3-class
df_3_class = df_master.copy()
df_2_class = df_master.copy()
df_2_class['Label_Binary'] = df_2_class['Label_Original'].apply(lambda x: 1 if x == 2 else 0)
```

```python
# =====================================================================
# CELL 4: XÂY DỰNG KIẾN TRÚC SELF-CARE (PHASE 2 & 3) - ĐÃ CẢI TIẾN
# =====================================================================
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import numpy as np

class SELFCARE_Wrist:
    def __init__(self):
        # 1. Khởi tạo 3 mô hình chuyên khoa: Thêm class_weight để xử lý mất cân bằng
        self.wb1 = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, class_weight='balanced_subsample')
        self.wb2 = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, class_weight='balanced_subsample')
        self.wb3 = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, class_weight='balanced_subsample')

        # 2. Khởi tạo Gating Model (Người gác cổng) bằng Decision Tree
        self.gating_model = DecisionTreeClassifier(min_samples_split=20, random_state=42)

        # 3. Định nghĩa các cảm biến cho từng nhánh theo đúng bài báo
        self.features_wb1 = ['BVP_mean', 'BVP_std', 'BVP_rms', 'BVP_dyn_range',
                             'EDA_mean', 'EDA_std', 'EDA_min', 'EDA_max', 'EDA_dyn_range', 'EDA_slope',
                             'TEMP_mean', 'TEMP_std', 'TEMP_min', 'TEMP_max', 'TEMP_dyn_range', 'TEMP_slope']

        self.features_wb2 = ['ACC_mean', 'ACC_std', 'ACC_integral',
                             'BVP_mean', 'BVP_std', 'BVP_rms', 'BVP_dyn_range',
                             'EDA_mean', 'EDA_std', 'EDA_min', 'EDA_max', 'EDA_dyn_range', 'EDA_slope']

        self.features_wb3 = ['BVP_mean', 'BVP_std', 'BVP_rms', 'BVP_dyn_range',
                             'EDA_mean', 'EDA_std', 'EDA_min', 'EDA_max', 'EDA_dyn_range', 'EDA_slope']

        # Gating model chỉ nhìn vào thông số ACC (Chuyển động tay)
        self.acc_features = ['ACC_mean', 'ACC_std', 'ACC_integral']

    def fit_system(self, X_train, y_train):
        # Bước A: Huấn luyện 3 nhánh trên dữ liệu Train
        self.wb1.fit(X_train[self.features_wb1], y_train)
        self.wb2.fit(X_train[self.features_wb2], y_train)
        self.wb3.fit(X_train[self.features_wb3], y_train)

        # Bước B: Sinh nhãn cho Gating Model (Tính Training Loss)
        prob_wb1 = self.wb1.predict_proba(X_train[self.features_wb1])
        prob_wb2 = self.wb2.predict_proba(X_train[self.features_wb2])
        prob_wb3 = self.wb3.predict_proba(X_train[self.features_wb3])

        classes = list(self.wb1.classes_)
        indices = [classes.index(lbl) for lbl in y_train]

        loss_wb1 = [-np.log(max(p[idx], 1e-10)) for p, idx in zip(prob_wb1, indices)]
        loss_wb2 = [-np.log(max(p[idx], 1e-10)) for p, idx in zip(prob_wb2, indices)]
        loss_wb3 = [-np.log(max(p[idx], 1e-10)) for p, idx in zip(prob_wb3, indices)]

        losses = np.vstack((loss_wb1, loss_wb2, loss_wb3)).T
        gating_labels = np.argmin(losses, axis=1)

        # Bước C: Dạy Người gác cổng
        self.gating_model.fit(X_train[self.acc_features], gating_labels)

    def predict_system(self, X_test):
        selected_branches = self.gating_model.predict(X_test[self.acc_features])
        pred_wb1 = self.wb1.predict(X_test[self.features_wb1])
        pred_wb2 = self.wb2.predict(X_test[self.features_wb2])
        pred_wb3 = self.wb3.predict(X_test[self.features_wb3])

        final_predictions = []
        for i, branch in enumerate(selected_branches):
            if branch == 0: final_predictions.append(pred_wb1[i])
            elif branch == 1: final_predictions.append(pred_wb2[i])
            else: final_predictions.append(pred_wb3[i])

        return np.array(final_predictions)

print("Đã thiết lập xong Cấu trúc Đa chuyên gia (Cải tiến Weight)!")
```

```python
# =====================================================================
# CELL 5: THỰC THI PHASE 4 - ĐÁNH GIÁ CHÉO (LOSO-CV) & LỌC THỜI GIAN
# =====================================================================
import numpy as np
from sklearn.metrics import accuracy_score, f1_score
from scipy import stats

def temporal_smoothing(predictions, window_size=12):
    smoothed_preds = []
    for i in range(len(predictions)):
        start_idx = max(0, i - window_size + 1)
        window_history = predictions[start_idx : i + 1]
        majority_vote = stats.mode(window_history, keepdims=True)[0][0]
        smoothed_preds.append(majority_vote)
    return np.array(smoothed_preds)

def evaluate_selfcare_loso(df, label_col, task_name):
    print(f"\n" + "="*55)
    print(f" BẮT ĐẦU CHẠY LOSO-CV CHO: {task_name}")
    print("="*55)

    subjects = df['Subject_ID'].unique()
    accuracies = []
    f1_scores = []

    for test_sub in subjects:
        train_data = df[df['Subject_ID'] != test_sub]
        test_data = df[df['Subject_ID'] == test_sub]

        X_train = train_data.drop(columns=['Subject_ID', 'Label_Original', 'Label_Binary'], errors='ignore')
        y_train = train_data[label_col].values

        X_test = test_data.drop(columns=['Subject_ID', 'Label_Original', 'Label_Binary'], errors='ignore')
        y_test = test_data[label_col].values

        model = SELFCARE_Wrist()
        model.fit_system(X_train, y_train)
        raw_predictions = model.predict_system(X_test)

        # BƯỚC C: Áp dụng Late Fusion (Lọc thời gian 60 giây = 12 bước trượt 5s)
        final_predictions = temporal_smoothing(raw_predictions, window_size=12)

        acc = accuracy_score(y_test, final_predictions)
        f1 = f1_score(y_test, final_predictions, average='macro', zero_division=0)

        accuracies.append(acc)
        f1_scores.append(f1)
        print(f"  -> Test trên {test_sub:^4} | Accuracy = {acc*100:.2f}% | Macro F1 = {f1*100:.2f}%")

    print("-" * 55)
    print(f" KẾT QUẢ TỔNG THỂ CỦA KIẾN TRÚC SELF-CARE:")
    print(f"   + Accuracy trung bình : {np.mean(accuracies)*100:.2f}%")
    print(f"   + Macro F1 trung bình : {np.mean(f1_scores)*100:.2f}%")
    print("-" * 55)

# Kích hoạt
evaluate_selfcare_loso(df_2_class, 'Label_Binary', "TASK 2-CLASS (Stress vs Non-Stress)")
evaluate_selfcare_loso(df_3_class, 'Label_Original', "TASK 3-CLASS (Baseline vs Stress vs Amusement)")

print("HOÀN THÀNH KẾT QUẢ THEO SELF-CARE CẢI TIẾN!")
```
